```python
data_feat=data.iloc[:,4:17]
data_feat.head()
```

```python
import warnings
warnings.filterwarnings('ignore')
for feature in data_feat:
    Q1=data_feat[feature].quantile(0.25)
    Q3=data_feat[feature].quantile(0.75)
    IQR=Q3-Q1
    lower=Q1-1.5*IQR
    upper=Q3+1.5*IQR
    data_feat[feature][(data_feat[feature]<lower)]=lower
    data_feat[feature][(data_feat[feature]>upper)]=upper
```

```python
for feature in data_feat:
    sn.boxplot(x=data_feat[feature])
    plt.show()
```

```python
from sklearn import preprocessing
scaler=preprocessing.StandardScaler()
data_feat=pd.DataFrame(scaler.fit_transform(data_feat),columns=data_feat.columns)
```

```python
data_meta=data.iloc[:,0:4]
data=pd.concat([data_meta,data_feat],axis=1)
data.head()
```

```python
corr=data_feat.corr()
plt.figure(figsize=(12,8))
sn.heatmap(corr, annot=True).set(title='Coorelations Between Audio Features');
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
vif = pd.DataFrame()
vif["variables"] = data_feat.columns
vif["VIF"] = [variance_inflation_factor(data_feat.values, i) for i in range(data_feat.shape[1])]
vif
```

```python
from sklearn.decomposition import PCA
pca=PCA().fit(data_feat)
plt.plot(pca.explained_variance_ratio_.cumsum(),marker='o',linestyle='--')
plt.title('Explained Variance by Components')
plt.xlabel('No of Components')
plt.ylabel('Cumaltive Explained Variance')
plt.axhline(0.9,c='r')
plt.axvline(8,c='g');
```
